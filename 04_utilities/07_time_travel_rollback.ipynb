```python
# -----------------------------------------
# VIEW DELTA TABLE HISTORY (AUDIT LOG)
# -----------------------------------------

fact_history = spark.sql("DESCRIBE HISTORY real_time_projects.ecommerce_historical.fact_sales")

display(fact_history)

# -----------------------------------------
# QUERY PREVIOUS VERSIONS (TIME TRAVEL)
# -----------------------------------------
fact_version =spark.read \
    .format("delta") \
    .option("versionAsOf", 1) \
    .table("real_time_projects.ecommerce_historical.fact_sales")

display(fact_version)
```

```python
%sql
-----------------------------------
-- SIMULATE AN INCORRECT UPDATE
-----------------------------------

----------------------------------
-- STEP 1 :MAKE A MISTAKE
----------------------------------

-- 1. Wrong update
UPDATE real_time_projects.ecommerce_historical.fact_sales
SET revenue = 0;

-- 2. Validate the mistake
SELECT SUM(revenue) FROM real_time_projects.ecommerce_historical.fact_sales;
```

```python
%sql
-------------------------------------------
-- Step 2 :RESTORE TO A PREVIOUS VERSION
-------------------------------------------

-- 1. Find correct version
DESCRIBE HISTORY real_time_projects.ecommerce_historical.fact_sales;

-- 2: Restore
RESTORE TABLE real_time_projects.ecommerce_historical.fact_sales TO VERSION AS OF 1;
```

```python
%sql
-----------------------------------------
-- Step 3:VALIDATE SUCESSFULL ROLLBACK
-----------------------------------------

-- 1: Revenue check
SELECT SUM(revenue) FROM real_time_projects.ecommerce_historical.fact_sales;

-- 2: Check history again
DESCRIBE HISTORY real_time_projects.ecommerce_historical.fact_sales;
```
